```python
import numpy as np
import random
import math as m
import timeit
import matplotlib.pyplot as plt
import scipy as sp
import scipy.special

def busquedaBinaria(lista, nroBuscado, izquierda, derecha):
    if izquierda > derecha:
        return -1
    indiceDelElementoDelMedio = (izquierda + derecha) // 2
    elementoDelMedio = lista[indiceDelElementoDelMedio]
    if elementoDelMedio == nroBuscado:
        return indiceDelElementoDelMedio
    if nroBuscado < elementoDelMedio:
        return busquedaBinaria(lista, nroBuscado, izquierda, indiceDelElementoDelMedio - 1)
    else:
        return busquedaBinaria(lista, nroBuscado, indiceDelElementoDelMedio + 1, derecha)
		


def busquedaTernaria(lista, nroBuscado, indiceIzquierda, indiceDerecha):             
    if (indiceDerecha >= indiceIzquierda):                                             
        mid_index1 = indiceIzquierda + (indiceDerecha - indiceIzquierda) // 3                
        mid_index2 = indiceDerecha - (indiceDerecha - indiceIzquierda) // 3              
        if (lista[mid_index1] == nroBuscado):                               
            return mid_index1                                                  
        if (lista[mid_index2] == nroBuscado):                               
            return mid_index2                                                  
                                                                                
        if (nroBuscado < lista[mid_index1]):                                
            return busquedaTernaria(lista, nroBuscado, indiceIzquierda, mid_index1-1)
        elif (nroBuscado > lista[mid_index2]):                              
            return busquedaTernaria(lista, nroBuscado, mid_index2+1, indiceDerecha)
        else:                                                                   
            return busquedaTernaria(lista, nroBuscado, mid_index1+1, mid_index2-1)
    return -1;  


n_min=1
n_max=2000
iteraciones = 2
st=10

t_busqueda_binaria = list()
for n in range(n_min, n_max, st):
    suma = 0
    v = random.sample(range(0, n), 1)[0]
    for foo in range(1,iteraciones):
        suma += timeit.timeit(lambda: busquedaBinaria(list(range(1,n)),n, 0, len(list(range(1,n))) -1), number = 1)
    promedio = suma / iteraciones
    t_busqueda_binaria.append(promedio)
    
t_busqueda_ternaria = list()
for n in range(n_min, n_max, st):
    suma = 0
    v = random.sample(range(0, n), 1)[0]
    for foo in range(1,iteraciones):
        suma += timeit.timeit(lambda: busquedaBinaria(list(range(1,n)),n, 0, len(list(range(1,n))) -1), number = 1)
    promedio = suma / iteraciones
    t_busqueda_ternaria.append(promedio)

eje_x = list(range(n_min, n_max, st))
plt.plot(eje_x, t_busqueda_binaria, label = 'Búsqueda binaria')
plt.plot(eje_x, t_busqueda_ternaria, label = 'Búsqueda ternaria')
plt.xlabel('n')
plt.ylabel('costo')

plt.legend()
```
